**Customizing the visualization**

```python
--These lines of code are used to create a bar graph. The first line is used to import the frame of the graph. the next liens are used to add details to those lines, such as the x and y axis, the color of the bar, the lines used the frmae the bar and the name of the of the bar graph.
```

```python
import matplotlib.pyplot as plt
df['age'].hist(color='crimson', edgecolor='black', bins=5)
plt.title('Age Distribution among NHANES Survey Respondents')
plt.xlabel('Age (years)')
plt.ylabel('Number of Respondents')
plt.grid(False)
```

**Seaborn Library**

```python
--These lines are used to create a violin graph. In the viloin the we have created the plot to have the x label show the diabetes status and the y label show the BMI and based off of the BMI determines whether or not you have diabetes
```

```python
import seaborn as sns
```

```python
# Filter to only Yes (1) and No (2) diabetes responses
df_diabetes = df[df['has_diabetes'].isin([1, 2])].copy()

diabetes_map={1:'Yes', 2:'No'}
df_diabetes['has_diabetes']=df_diabetes['has_diabetes'].map(diabetes_map)

# Create violin plot
sns.violinplot(data=df_diabetes, x='has_diabetes', y='bmi')
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes Status (1=Yes, 2=No)')
plt.ylabel('BMI (kg/m)')
```

```python
--These lines of code create a heat map. Nased on the age, bmi, systolic_bp, hba1c, and fasting glucose, it will show how deep of a red color it is.
```

```python
# Select only numeric columns
numeric_cols = df[['age', 'bmi', 'systolic_bp', 'hba1c', 'fasting_glucose']].corr()

sns.heatmap(numeric_cols, annot=True, cmap='coolwarm', center=0)
plt.title('Correclation Between Numeric Varaibles')
```
