```
def c_to_f(c_temp):
  f_temp=(c_temp * (9/5)) + 32
  return f_temp
```

```
f_temp = c_to_f(4)
print(f_temp)
```

**Calculating BMI**

```
//these lines of code are used to calculate the BMI
```

```
def calculate_BMI(height_m, weight_kg):
  BMI=weight_kg/(height_m**2)
  return BMI
```

```
calculate_BMI(weight_kg=75,height_m=1.7)
```
