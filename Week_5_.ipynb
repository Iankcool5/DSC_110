**Transform Variables**

```python
--This line of code creates a system where depending on the bmi of the individual, it'll give thme the response of whether theey are any of the 4 options in the labels lines
```

```python
df['bmi_category'] = pd.cut(df['bmi'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

```python
df['bmi_category'].head()
```

```python
df['bmi_category'].value_counts()
```

**Categorical Variable Transformation**

```python
df['education'].head()
```

```python
--This line of code creates a map of statemenst to give based on what the user gives for their level of education
```

```python
education_map = {
    1: 'Less than 9th grade',
    2: '9-11th grade',
    3: 'High school grad',
    4: 'Some college or AA degree',
    5: 'College graduate',
    9: 'Do not know',
    'nan': 'Missing'
}
```

```python
education_map[1]
```

```python
--This shows the number of indivaiduals in each category
```

```python
df['education_label'] = df['education'].map(education_map)
df['education_label'].value_counts()
```
